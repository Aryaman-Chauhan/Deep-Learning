```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.cvtColor(cv2.imread(r"C:\Users\asus\Downloads\Image%20(19).jpg"), cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
image_rect = img.copy()
cv2.rectangle(image_rect, (876,27), (962,169), (0,0,255), thickness=5, lineType = cv2.LINE_AA)
cv2.rectangle(image_rect, (1499,150), (1639,262), (0,0,255), thickness=5, lineType = cv2.LINE_AA)
cv2.rectangle(image_rect, (1529, 377), (1654, 484), (0,0,255), thickness=5, lineType = cv2.LINE_AA)
cv2.rectangle(image_rect, (1025, 320), (1116, 494), (0,0,255), thickness=5, lineType = cv2.LINE_AA)
cv2.rectangle(image_rect, (1230, 512), (1355, 599), (0,0,255), thickness=5, lineType = cv2.LINE_AA)
plt.imshow(image_rect)
plt.show()
```
